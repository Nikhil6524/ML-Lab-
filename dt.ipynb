```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('training.xlsx')
df.head()
```

```python
x=df.drop('output',axis=1)
y=df['output']
```

```python
def custom_round(value):
    integer_part = int(value)
    decimal_part = value - integer_part
    if decimal_part == 0.5:
        return value
    else:
        return round(value)

df['output'] = df['output'].apply(custom_round)
df['output'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth':list(np.arange(20)),  # Maximum depth of the tree
    'min_samples_split': list(np.arange(20)),   # Minimum number of samples required to split an internal node
    'min_samples_leaf': list(np.arange(20)),     # Minimum number of samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2', None]  # Number of features to consider for splitting
}
```

```python
gk=GridSearchCV(dt,param_grid=param_grid)
```
